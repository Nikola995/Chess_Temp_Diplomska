## Helper functions

```python
def filter_by(files, num_users = 0, user = 0):
    if num_users:
        filtered_files = [x for x in files if int(x.split('_')[1]) == num_users]
        # Sort ascending
        filtered_files.sort(key=lambda x: int(x.split('_')[2]),
                                  reverse=True)
        return filtered_files
    elif user:
        filtered_files = [x for x  in files if int(x.split('_')[2]) == user]
        # Sort ascending
        filtered_files.sort(key=lambda x: int(x.split('_')[1]))
        return filtered_files
    else:
        raise ValueError()
#first_user = filter_by(predictions, user=400)

def round_down(x, base=5):
    return x - (x % base)

def round_up(x, base=5):
    result = x + ((base - x) % base)
    # In this perticular case hard cap to 100
    # because we are working with accuracy
    if result > 100:
       result = 100
    return result

# print(round_down(df.describe()['min'],5))
# print(round_up(df.describe()['max'],5))
```

```python
def unique_predictions(pred_folder, pred_file):
    pred_df = pd.read_csv(os.path.join(pred_folder,
                                       pred_file),
                          index_col=None)
    # %% Train data
    train_datasets = os.listdir(os.path.join(os.getcwd(),
                                             'dt_chess',
                                             'datasets'))
    train_data = [x for x in train_datasets if x.partition('_')[0] == pred_file.split('_')[2]][0]
    train_df = pd.read_csv(os.path.join(os.getcwd(),
                                        'dt_chess',
                                        'datasets',
                                        train_data),
                           index_col=(0))
    train_df.rename(columns={'positions': 'board', 'moves': 'move'}, inplace=True)
    train_df.drop_duplicates(inplace=True)
    # Remove duplicates from train data
    pred_df_fin = pd.merge(pred_df, train_df,
                           on=["board","move"], how='left',
                           indicator='Exist')
    #common_total = len(pred_df_fin[pred_df_fin['Exist'] == 'both'])
    pred_df_fin = pred_df_fin[pred_df_fin['Exist'] == 'left_only']
    return pred_df_fin

def get_stats_for_predictions(pred_all_1_df):
    total = len(pred_all_1_df)
    fp = len(pred_all_1_df[pred_all_1_df['illegal_top_move']])
    tp1 = len(pred_all_1_df[pred_all_1_df['move'] == pred_all_1_df['engine_move_1']])
    tp2 = len(pred_all_1_df[pred_all_1_df['move'] == pred_all_1_df['engine_move_2']])
    tp3 = len(pred_all_1_df[pred_all_1_df['move'] == pred_all_1_df['engine_move_3']])
    tn1 = total - tp1
    tn2 = total - (tp1 + tp2)
    tn3 = total - (tp1 + tp2 + tp3)
    acc1 = (tp1/total)*100
    acc2 = (tp2/total)*100
    acc3 = (tp3/total)*100
    acctop2 = ((tp1 + tp2)/total)*100
    acctop3 = ((tp1 + tp2 + tp3)/total)*100
    return (total, fp, tp1, tp2, tp3, tn1, tn2, tn3, acc1, acc2, acc3, acctop2, acctop3)

def get_stats_for_all_predictions(pred_folder, user_predictions,
                                  unique=True, verbose=True):
    stats = []
    pred_list = tqdm(user_predictions) if verbose else user_predictions
    for user_pred_file in pred_list:
        if unique:
            pred_df = unique_predictions(pred_folder, user_pred_file)
        else:
            pred_df = pd.read_csv(os.path.join(pred_folder,
                                               user_pred_file),
                                  index_col=None)
        stats.append(get_stats_for_predictions(pred_df))
    return pd.DataFrame(stats, columns =["total", "fp",
                                         "tp1", "tp2", "tp3",
                                         "tn1", "tn2", "tn3",
                                         "acc1", "acc2", "acc3", "acctop2", "acctop3"])

def get_stats_for_predictions_by_turn(pred_folder, user_predictions, verbose = True):
    stats_by_turn_all_files = {}
    pred_list = tqdm(user_predictions) if verbose else user_predictions
    for user_pred_file in pred_list:
        pred_df = unique_predictions(pred_folder, user_pred_file)
        pred_by_turn = pred_df.groupby(['turn'])
        pred_single_turn_dfs = [pred_by_turn.get_group(x) for x in pred_by_turn.groups]
        pred_turn_stats = [get_stats_for_predictions(x) for x in pred_single_turn_dfs]
        pred_turn_stats_df = pd.DataFrame(pred_turn_stats, columns =["total", "fp",
                                         "tp1", "tp2", "tp3",
                                         "tn1", "tn2", "tn3",
                                         "acc1", "acc2", "acc3", "acctop2", "acctop3"])
        pred_turn_stats_df.index = pred_by_turn.groups.keys()
        stats_by_turn_all_files[user_pred_file[:-len('.csv')]] = pred_turn_stats_df
    return stats_by_turn_all_files

def get_train_losses(losses_dir, loss_files):
    losses_df = []
    for loss_file in loss_files:
        loss_df = pd.read_csv(os.path.join(losses_dir, loss_file),
                              header=0)
        loss_epoch = loss_file.rpartition('_')[-1][:-len('.csv')]
        loss_df.rename(columns={loss_df.columns[0]: f"epoch_{loss_epoch}"},
                      inplace=True)
        losses_df.append(loss_df)
    return pd.concat(losses_df,axis=1)
```

```python
### All predictions plot ##
def plot_accuracy_by_epoch(accuracy_df, loss_type, state_emb_type):
    #PDF plot
    plt.figure(figsize=(16, 10))
    ax = sns.lineplot(data = accuracy_df)
    ax.set(xlabel='Epoch')
    ax.set(ylabel='Accuracy')
    #Set major and minor ticks
    ax.yaxis.set_major_locator(MultipleLocator(5))
    ax.yaxis.set_minor_locator(MultipleLocator(1))
    ax.tick_params(axis='y', which='both', left=True)
    ax.xaxis.set_major_locator(MultipleLocator(1))
    #ax.xaxis.set_minor_locator(MultipleLocator(1))
    ax.tick_params(axis='x', which='both', bottom=True)
    ax.tick_params(axis="x", which="both", rotation=45)
    #Extend y-limits
    #Get the min of all accuracy values
    plt.ylim(round_down(accuracy_df.min().min(),5),
             round_up(accuracy_df.max().max(),5))
    #Tighted x-limits
    plt.xlim(accuracy_df.index.min(),
             accuracy_df.index.max())
    #Set grid
    plt.grid(visible = True, which = 'both', axis = 'y', linestyle = '-',zorder=0)
    plt.grid(visible = True, which = 'both', axis = 'x', linestyle = '-',zorder=0)
    plt.title(f"All predictions accuracies by epoch (Training loss - {loss_type}) "
              f"({state_emb_type} model)")
    plt.tight_layout()
    plt.show()

def plot_acc_by_turn(accuracy_df, epoch, loss_type, state_emb_type):
    #PDF plot
    plt.figure(figsize=(16, 10))
    ax = sns.lineplot(data = accuracy_df)
    ax.set(xlabel='Turn')
    ax.set(ylabel='Accuracy')
    #Set major and minor ticks
    ax.yaxis.set_major_locator(MultipleLocator(5))
    ax.yaxis.set_minor_locator(MultipleLocator(1))
    ax.tick_params(axis='y', which='both', left=True)
    ax.xaxis.set_major_locator(MultipleLocator(5))
    ax.xaxis.set_minor_locator(MultipleLocator(1))
    ax.tick_params(axis='x', which='both', bottom=True)
    ax.tick_params(axis="x", which="both", rotation=45)
    #Extend y-limits
    #Get the min of all accuracy values
    global_min = accuracy_df.describe().loc['min'].min()
    global_max = accuracy_df.describe().loc['max'].max()
    plt.ylim(round_down(global_min,5),
             round_up(global_max,5))
    #Tighted x-limits
    plt.xlim(accuracy_df.index.min(),
             accuracy_df.index.max())
    #Set grid
    plt.grid(visible = True, which = 'both', axis = 'y', linestyle = '-',zorder=0)
    plt.grid(visible = True, which = 'both', axis = 'x', linestyle = '-',zorder=0)
    plt.title(f"All predictions accuracy by turn ({epoch}) "
              f"(Training loss - {loss_type}) ({state_emb_type} model)")
    plt.tight_layout()
    plt.show()
    
def plot_train_loss(losses_df, epoch, loss_type, state_emb_type):
    #PDF plot
    plt.figure(figsize=(16, 10))
    ax = sns.lineplot(data = losses_df, alpha  = 0.7)
    ax.set(xlabel='Iteration')
    ax.set(ylabel='Train Loss')
    #Set major and minor ticks
    ax.yaxis.set_major_locator(MultipleLocator(0.5))
    ax.yaxis.set_minor_locator(MultipleLocator(0.1))
    ax.tick_params(axis='y', which='both', left=True)
    ax.xaxis.set_major_locator(MultipleLocator(50))
    ax.xaxis.set_minor_locator(MultipleLocator(10))
    ax.tick_params(axis='x', which='both', bottom=True)
    ax.tick_params(axis="x", which="both", rotation=45)
    #Extend y-limits
    plt.ylim(round_down(losses_df.min().min(),0.5),
             round_up(losses_df.max().max(),0.5))
    #Tighted x-limits
    plt.xlim(0,len(loss_df)-1)
    #Set grid
    plt.grid(visible = True, which = 'major', axis = 'y', linestyle = '-',zorder=0)
    plt.grid(visible = True, which = 'major', axis = 'x', linestyle = '-',zorder=0)
    plt.title(f"First model's train loss for {epoch} (batch size = 256)"
              f" (Training loss - {loss_type}) ({state_emb_type} model)")
    plt.tight_layout()
    plt.show()
```
